# Proof Of Concept ML modeling

```python
import pandas as pd
import numpy as np


df = pd.read_csv('../data/flats.csv')
df.head()
```

```python
df.describe()
```

```python
batches = np.array_split(df, 5)

df = pd.DataFrame(batches[0])
```

```python
X = df[['floor_area_sqm', 'lease_commence_date']]
y = df['resale_price']

X.shape
```

```python
from sklearn.linear_model import LinearRegression # base
from sklearn.linear_model import SGDRegressor # more complex with regularization
from sklearn.linear_model import RANSACRegressor # for outliers



lr = LinearRegression()
sdg = SGDRegressor()
ransac = RANSACRegressor()
```

```python
lr.fit(X, y)
ransac.fit(X, y)
```

```python
print('Linear regression mean accuracy:', lr.score(X, y))
print('RANSAC regression mean accuracy:', ransac.score(X, y))
```

### Conclusion:

In this proof-of-concept study, we trained a basic machine learning model using only two numerical features and without implementing any preprocessing or parameter tuning. Our initial results yielded accuracies of 0.43 and 0.13, indicating room for substantial improvement.

They serve as a baseline, highlighting the model's potential and confirming that machine learning can be applied to our project.
